```python
# importing python libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
# import csv file
df = pd.read_csv('Amazon Sales data.csv', encoding= 'unicode_escape')
```

```python
df.head()
```

```python
#Check Duplicates
df['Order ID'].duplicated()
```

```python
df.drop_duplicates('Order ID')
```

```python
df.describe()
```

```python
# Assuming 'df' to  DataFrame
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df.describe()
```

# Exploratory Data Analysis

```python
#Most of The Order Place Year
sns.histplot(x='Order Date',data=df,kde=True,color='Red')
```

```python
#Online and offline Total Revenue
sns.histplot(x='Total Revenue',data=df,hue='Sales Channel',element='poly')
plt.grid()
```

```python
#Total Units Sold By Region
sns.barplot(y='Region',x='Units Sold',data=df,palette='CMRmap')
```

```python
#Total Profit By Item Type
sns.histplot(x='Total Profit',hue='Item Type',data=df)
```

```python
#Total Profit by Order Priority
sns.boxplot(y='Total Profit',data=df,x='Order Priority',color='Orange',linewidth=0.5)
plt.grid()
```

```python
#Unit Cost By Item Type
sns.barplot(x='Unit Cost',y='Item Type',data=df,palette='CMRmap')
```

```python
#Order Year By Region
sns.histplot(y='Order Date',hue='Region',data=df)
```

```python
#Year on Year Growth 2010 to 2017
sns.lineplot(x='Order Date',y='Total Profit',data=df,color='Green')
sns.set(rc={'figure.figsize':(15,5)})
plt.xlabel('Order Year',fontsize=25)
plt.ylabel('Total Profit',fontsize=25)
plt.title('Year on Year Growth',fontsize=30)
```

```python
#Total Profit By Region
sns.barplot(y='Total Profit',x='Region',data=df,palette='cividis',errorbar=None)
```
